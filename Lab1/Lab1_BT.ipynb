2.VIẾT CHƯƠNG TRÌNH NẠP MỘT ẢNH VÀ HOÁN ĐỔI GIÁ TRỊ CÁC MÀU LƯU CÁC ẢNH VÀO MÁY

```python
#MÀU 1
import colorsys
import imageio.v2 as iio
import matplotlib.pylab as plt
rgb = iio.imread('AMONGUS.jpg')
rgb2hsv = np.vectorize (colorsys.rgb_to_hsv)
h, s, v = rgb2hsv (rgb [:,:,0]/255., rgb [:,:,1]/255., rgb[:,:,2]/255.)
h *= h
hsv2rgb = np.vectorize (colorsys.hsv_to_rgb)
rgb2 = hsv2rgb (h, s, v)
rgb2 = np.array(rgb2).transpose((1, 2, 0))
rgb2_uint8 = np.clip(rgb2 * 255, 0, 255).astype(np.uint8)
iio.imsave('AMONGUS_CL1.jpg', rgb2_uint8)
plt.imshow(rgb2_uint8)
plt.show()
```

```python
#MÀU 2
rgb = iio.imread('AMONGUS.jpg')
rgb2hsv = np.vectorize (colorsys.rgb_to_hsv)
h, s, v = rgb2hsv (rgb [:,:,1]/255., rgb [:,:,2]/255., rgb[:,:,1]/255.)
h *= h
hsv2rgb = np.vectorize (colorsys.hsv_to_rgb)
rgb2 = hsv2rgb (h, s, v)
rgb2 = np.array(rgb2).transpose((1, 2, 0))
rgb2_uint8 = np.clip(rgb2 * 255, 0, 255).astype(np.uint8)
iio.imsave('AMONGUS_CL2.jpg', rgb2_uint8)
plt.imshow(rgb2_uint8)
plt.show()
```

```python
import scipy.ndimage as sn
import imageio.v2 as iio
import numpy as np
import matplotlib.pylab as plt
a = iio.imread('baby.jpeg', mode='F')
k = np.ones((5,5))/25
b = sn.convolve(a, k) .astype (np. uint8)
iio.imsave('baby_mean_filter.png', b)
print (b)
plt.imshow (b)
plt.show()
```
